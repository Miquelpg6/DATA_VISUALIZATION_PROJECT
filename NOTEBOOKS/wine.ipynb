```python
import pandas as pd
import numpy as np
import pymysql
import sqlalchemy as alch
import os

from dotenv import load_dotenv
from getpass import getpass
```

```python
# Read CSV and append the column Variety "Red"

red = pd.read_csv("../DATA/Red.csv", encoding='utf-8')
red["Variety"] = "Red"
red
```

```python
# Read the csv White and append the column Variety "White"

white = pd.read_csv("../DATA/White.csv", encoding='utf-8')
white["Variety"] = "White"
white
```

```python
# Concat the two DF into one

wine = pd.concat([red, white], ignore_index=True, )
wine
```

```python
wine.to_csv('../DATA/wine.csv', index=False, encoding='utf-8')
```

```python
token = getpass()
```

```python
dbname = "wine"
sql_password = token
connectionData = f"mysql+pymysql://root:{sql_password}@localhost/{dbname}"
engine = alch.create_engine(connectionData)
wine.to_sql("WINE", if_exists="append", con=engine, index=False)
```
